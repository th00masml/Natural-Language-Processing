```python
import pandas as pd
import re
import string
import nltk
pd.set_option('display.max_colwidth', 100)

stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()

data = pd.read_csv("SMSSpamCollection.tsv", sep='\t')
data.columns = ['label', 'body_text']
```

```python
def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [ps.stem(word) for word in tokens if word not in stopwords]
    return text
```

```python
from sklearn.feature_extraction.text import CountVectorizer # Vectorize libray from sciklearn

count_vect = CountVectorizer(analyzer=clean_text) # Assign the function
X_counts = count_vect.fit_transform(data['body_text']) # Train the data
```

```python
data_sample = data[0:20]

count_vect_sample = CountVectorizer(analyzer=clean_text) # Assign the function
X_counts_sample = count_vect_sample.fit_transform(data_sample['body_text']) # Train the data
print(X_counts_sample.shape) # Check the shape of sample data
print(count_vect_sample.get_feature_names()) # Print all of unique words(columns) in sample dataset
```
