```python
import pandas as pd
import re
import string
import nltk
pd.set_option('display.max_colwidth', 100)

stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()

data = pd.read_csv("SMSSpamCollection.tsv", sep='\t')
data.columns = ['label', 'body_text']
```

```python
def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = " ".join([ps.stem(word) for word in tokens if word not in stopwords])
    return text

data['cleaned_text'] = data['body_text'].apply(lambda x: clean_text(x))
data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
ngram_vect = CountVectorizer(ngram_range = (2,2)) # Range of n-grams. Here whe have only bigrams
X_counts = ngram_vect.fit_transform(data['cleaned_text']) # Train the data
```

```python
data_sample = data[0:20]

ngram_vect_sample = CountVectorizer(ngram_range = (2,2)) # Range of n-grams. Here whe have only bigrams
X_counts_sample = ngram_vect_sample.fit_transform(data_sample['cleaned_text']) # Train the data
print (X_counts_sample.shape) # Check out the shape of data
print(ngram_vect_sample.get_feature_names()) # Print all of unique 2 word combinations
```

```python
X_counts_df = pd.DataFrame(X_counts_sample.toarray())
X_counts_df.columns = ngram_vect_sample.get_feature_names() # Show the column names
X_counts_df
```
